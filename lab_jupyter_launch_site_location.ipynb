```python
# TASK 3: Calculate the distances between a launch site to its proximities
import pandas as pd
from math import radians, cos, sin, asin, sqrt

# Example launch_sites_df
# launch_sites_df = pd.DataFrame({
#     'Launch Site': ['Site A', 'Site B'],
#     'Lat': [28.5623, 34.6321],
#     'Long': [-80.5774, -120.6108]
# })

# Haversine distance function
def haversine_distance(lat1, lon1, lat2, lon2):
    """Calculate distance between two points on Earth in km."""
    lat1, lon1, lat2, lon2 = map(radians, [lat1, lon1, lat2, lon2])
    dlat = lat2 - lat1
    dlon = lon2 - lon1
    a = sin(dlat/2)**2 + cos(lat1) * cos(lat2) * sin(dlon/2)**2
    c = 2 * asin(sqrt(a))
    r = 6371  # Earth's radius in km
    return c * r

# Proximity coordinates
equator_lat = 0.0  # Equator latitude
coastline_coord = [28.5623, -80.5774]  # Example coastline coordinate (adjust as needed)

# Calculate distances
launch_sites_df['Distance_to_Equator_km'] = launch_sites_df.apply(
    lambda row: haversine_distance(row['Lat'], row['Long'], equator_lat, row['Long']),
    axis=1
)

launch_sites_df['Distance_to_Coast_km'] = launch_sites_df.apply(
    lambda row: haversine_distance(row['Lat'], row['Long'], coastline_coord[0], coastline_coord[1]),
    axis=1
)

# Optional: distance to nearest other launch site
def nearest_site_distance(lat, lon, df, current_site):
    distances = df[df['Launch Site'] != current_site].apply(
        lambda row: haversine_distance(lat, lon, row['Lat'], row['Long']),
        axis=1
    )
    return distances.min() if not distances.empty else 0

launch_sites_df['Distance_to_Nearest_LaunchSite_km'] = launch_sites_df.apply(
    lambda row: nearest_site_distance(row['Lat'], row['Long'], launch_sites_df, row['Launch Site']),
    axis=1
)

# Show final DataFrame
launch_sites_df

```

```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```
